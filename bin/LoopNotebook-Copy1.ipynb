```python
%load_ext autoreload
%autoreload 2
    
import sys
sys.path.append("../modules")

from loop import *

INO=Arduino('COM3')
HP=HP4155('GPIB0::17', debug=False)

path=f'C:/Users/Zucchi/Documents/Medidas/2406 - TN3 DP1 DN1 CB1/'

HP.SetIntTime("MED")
HP.SetStop("COMP")
```

```python
%matplotlib agg
%matplotlib agg

freq=10
count=48
    
for i in range(count):
    clear_output(wait=True)

    prog_bar=display('',display_id=True)
    
    for j in range(720):
        if datetime.datetime.now().time().minute%freq == 0: break        
        
        prog_bar.update(datetime.datetime.now().strftime("%H:%M:%S"))
        
        sleep(1)

    
    start=datetime.datetime.now().strftime("%y%m%d %H%M%S")
    printf(f"Measurement: {i+1}", path)
    
    clear_output(wait=True)

    printf(f"Chn{2} - DP1", path)

    HP.SetStop("COMP")
    
    INO.opench(2)

    HP.SingleDiode(-1, 1.5, 0.01, SMUN='SMU1', SMUP='SMU2')
    Plot(HP.SingleSave(path+f'DP1/{start}.csv', timeout=30, real=True), 'Vf', 'If')

    printf(f"Chn{3} - DN1", path)

    INO.opench(3)

    HP.SingleDiode(-1, 1.35, 0.01, SMUN='SMU1', SMUP='SMU2')
    Plot(HP.SingleSave(path+f'DN1/{start}.csv', timeout=30, real=True), 'Vf', 'If')
    HP.SingleDiode(1.35, 1.6, 0.001, SMUN='SMU1', SMUP='SMU2')
    Plot(HP.SingleSave(path+f'DN1/{start}2.csv', timeout=30, real=True), 'Vf', 'If')

    printf(f"Chn{4} - CB1", path)
    
    INO.opench(4)

    HP.SetStop("OFF")
    
    HP.Set2P(-0.5e-3, 0.5e-3, 200, SMUN='SMU1', SMUP='SMU2', Comp=4)
    R12=Plot2P(HP.SingleSave(path+f'CB1/12 - {start}.csv', timeout=30))

    HP.Set2P(-0.5e-3, 0.5e-3, 200, SMUN='SMU1', SMUP='SMU3')
    R13=Plot2P(HP.SingleSave(path+f'CB1/13 - {start}.csv', timeout=30))

    HP.Set2P(-0.5e-3, 0.5e-3, 200, SMUN='SMU1', SMUP='SMU4')
    R14=Plot2P(HP.SingleSave(path+f'CB1/14 - {start}.csv', timeout=30))

    HP.Set2P(-0.5e-3, 0.5e-3, 200, SMUN='SMU3', SMUP='SMU4')
    R34=Plot2P(HP.SingleSave(path+f'CB1/34 - {start}.csv', timeout=30))

    printf(f"R12={format(R12, '.2f')}    R13={format(R13, '.2f')}    R14={format(R14, '.2f')}    R34={format(R34, '.2f')}    ", path)
    plt.close('all')
```
